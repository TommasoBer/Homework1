```python
# collections.Counter()

from collections import Counter
tot_scarpe = int(input())
taglie_scarpe = list(map(int,input().split()))
taglie_paia = Counter(taglie_scarpe)
guadagno = 0
if tot_scarpe != len(taglie_scarpe):
    print("C'Ã¨ un errore")
else:
    n_clienti = int(input())
    for i in range(0,n_clienti):
        taglia, prezzo = map(int,input().split())
        
        if taglia in taglie_paia.keys():
            if taglie_paia[taglia] == 0 :
                #print(taglie_paia[taglia])
                guadagno = guadagno + 0
                #print("baccala")
            else:
                #print(taglie_paia[taglia])
                x = taglie_paia[taglia]
                xa = x-1
                taglie_paia[taglia] = xa
                #print(taglie_paia[taglia])
                guadagno = guadagno + prezzo
        else:
            guadagno = guadagno + 0
        
    print(guadagno)
```

```python
# DefaultDict Tutorial

from collections import defaultdict
n, m = map(int,input().split())
lista_a = defaultdict(list)
#lista_a = dict(lista_a)
lista_b = []
for i in range (0,n):
    x = str(input())
    lista_a[x].append(i+1)
for j in range(0,m):
    y = str(input())
    lista_b.append(y)
for word in lista_b:
    if word in lista_a.keys():
        #print(lista_a[word])
        u = lista_a[word]
        a = " ".join(str(n) for n in u)
        print(a)
    else:
        print("-1")
```

```python
# Word Order

from collections import Counter
n_parole = int(input())
set1 = []
for i in range (0,n_parole):
    x = input()
    set1.append(x)
x = Counter(set1)
chiavi = x.keys()
print(len(chiavi))
val = x.values()
val = list(val)
value = [str(a) for a in val]
val = " ".join(value)
print(val)
```

```python
# Collections.deque()

from collections import deque
n_operazioni = int(input())
d = deque()
for i in range(0, n_operazioni):
    comando_val = list(map(str, input().split()))
    if len(comando_val) == 1:
        comando = comando_val[0]
        if comando == "pop":
            
            d.pop()
        elif comando =="popleft":
            
            d.popleft()
    else:
        comando = comando_val[0]
        valore = comando_val[1]
        if comando == "append" :
                d.append(valore)
        elif comando == "appendleft":
                d.appendleft(valore)
d = list(d)
result = " ".join(d)
print(result)
```
